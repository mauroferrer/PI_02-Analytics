#### Mapa de calor con las correlaciones de data3

```python
df_numericas = data3.select_dtypes(include=['number']) #seleccionamos las columnas numemricas solamente
correlaciones = df_numericas.corr().round(2) # hacemos la correlacion
plt.figure(figsize=(13,9))
sns.heatmap(data = correlaciones, annot=True)
plt.show()
```

#### Encontramos valores atipicos de data3

```python
Q1 = df_numericas.quantile(0.25) 
Q3 = df_numericas.quantile(0.75)
IQR = Q3 - Q1
lim_inferior = Q1 - 1.5 * IQR 
lim_superior = Q3 + 1.5 * IQR 
mascaras = (df_numericas < lim_inferior) | (df_numericas > lim_superior) 
valores_atipicos = df_numericas[mascaras]
valores_atipicos
```

```python
from wordcloud import WordCloud
# Calcular la frequencia de palabras
word_freq = data['name'].str.split(expand=True).stack().value_counts()

# Crear la nube de palabras
wordcloud = WordCloud(width=800, height= 400, background_color = 'white').generate_from_frequencies(word_freq)

# Visualizamos la nube de palabras
plt.figure(figsize = (10,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
from wordcloud import STOPWORDS
```

#### Creamos un analisis de sentimientos sobre la coumna reviews

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
```

```python
data1['sentimiento'] = data1['reviews'].apply(lambda texto: sid.polarity_scores(str(texto))['compound'] if isinstance(texto, str) else 0)
```
